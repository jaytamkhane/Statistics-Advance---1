Theory Questions: 

1) What is a random variable in probability theory ?
>  A random variable is a function that maps the outcomes of a random experiment to real numbers. It's a way to abstractly represent the outcomes of a random experiment.

2) What are the types of random variables ?
>  The two main families of random variable types are discrete and continuous. Discrete random variables can only take on integer values. Continuous random variables can take on decimal values.

3) What is the difference between discrete and continuous distributions ?
>  A discrete distribution is one in which the data can only take on certain values, for example integers. A continuous distribution is one in which data can take on any value within a specified range (which may be infinite).

4) What are probability distribution functions (PDF) ?
>  A Probability Distribution Function (PDF) is a mathematical function that describes the likelihood of different outcomes for a random variable in a probabilistic experiment. The specific meaning of a PDF depends on whether the random variable is discrete or continuous.

5) How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF) ?
>  A probability distribution function (PDF) describes the likelihood of a random variable taking on a specific value, while a cumulative distribution function (CDF) describes the likelihood of a random variable taking on a value less than or equal to a specific value.

6) What is a discrete uniform distribution ?
>  A discrete uniform distribution is a type of probability distribution where a finite number of outcomes are equally likely. In other words, each outcome has the same probability of occurring.

7) What are the key properties of a Bernoulli distribution ?
>  A Bernoulli distribution is a discrete probability distribution that describes a random experiment with exactly two possible outcomes: "success" (often denoted as 1) and "failure" (denoted as 0). The probability of success is p, and the probability of failure is 1−p. 

8) What is the binomial distribution, and how is it used in probability ?
>  The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent and identical trials of a Bernoulli experiment. Each trial has only two possible outcomes: success (with probability p) and failure (with probability 1−p).

9) What is the Poisson distribution and where is it applied ?
>  The Poisson distribution is a discrete probability distribution that models the number of events occurring in a fixed interval of time, space, or any other domain, under the following conditions:
a) Events occur independently: The occurrence of one event does not affect the probability of another.
b) Constant average rate (λ): The average number of events in the interval is fixed.
c) Rare events: The probability of more than one event occurring in an infinitesimally small interval is negligible.

10) What is a continuous uniform distribution ?
>   The continuous uniform distribution describes a situation where all outcomes in a given interval are equally likely. It is characterized by a constant probability density function (PDF) over a specified range [a, b].

11) What are the characteristics of a normal distribution ?
>   The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric around its mean.
a) Symmetry: The distribution is symmetric about its mean (μ) & the left and right sides of the curve are mirror images.
b) Bell Shape: The graph is bell-shaped, with a single peak at the mean.
c) Defined by Two Parameters: mean(μ) and Standard deviation (σ)
d) Empirical Rule (68-95-99.7 Rule)

12) What is the standard normal distribution, and why is it important ?
>   The standard normal distribution is a special case of the normal distribution where the mean (μ) is 0 and the standard deviation (σ) is 1. It is denoted by Z.
a) Z-Scores for Standardization: Any normal distribution can be transformed into the standard normal distribution using the formula, this process, called standardization.
b) Simplifies Probability Calculations: By converting to the standard normal distribution, probabilities can be calculated using a single table or function instead of dealing with multiple unique normal distributions.
c) Easy Interpretation: Z-scores represent how many standard deviations a value is from the mean.

13) What is the Central Limit Theorem (CLT), and why is it critical in statistics ?
>   The Central Limit Theorem (CLT) is a fundamental concept in statistics that states When a sufficiently large number of independent random samples are drawn from any population with a finite mean (μ) and finite variance (𝜎2), the sampling distribution of the sample mean will approximate a normal distribution, regardless of the population's original distribution.

14) How does the Central Limit Theorem relate to the normal distribution ?
>   The Central Limit Theorem (CLT) is closely related to the normal distribution because it explains why the sampling distribution of the sample mean tends to follow a normal distribution, even when the original population is not normally distributed.
a) Normal Approximation: The CLT states that, for large sample sizes n, the distribution of the sample mean will approximate a normal distribution, regardless of the shape of the population distribution.
b) Convergence to Normality: If the original population is already normally distributed, the sample means will also follow a normal distribution for any sample size.

15) What is the application of Z statistics in hypothesis testing ?
>   Z-statistics are widely used in hypothesis testing when the sampling distribution of the sample mean is approximately normal, typically when the population is known to be normally distributed or when the sample size is large enough (thanks to the Central Limit Theorem).Z-statistics, also known as z-scores, measure how many standard deviations an observation or sample mean is away from the population mean.

16) How do you calculate a Z-score, and what does it represent ?
>   A Z-score (also called a standard score) represents the number of standard deviations a data point (or sample mean) is away from the mean of the population or sample. In other words, it standardizes the data, allowing you to compare values from different distributions or scales.

17) What are point estimates and interval estimates in statistics ?
>   point estimates are single values that estimate a parameter, while interval estimates are ranges of values that estimate a parameter.

18) What is the significance of confidence intervals in statistical analysis ?
>   What does a confidence interval tell you? he confidence interval tells you more than just the possible range around the estimate. It also tells you about how stable the estimate is. A stable estimate is one that would be close to the same value if the survey were repeated.

19) What is the relationship between a Z-score and a confidence interval ?
>   The Z-score helps calculate the margin of error by the multiplying it with the standard error. This margin of error defines the range of the confidence interval around the sample mean.

20) How are Z-scores used to compare different distributions ?
>   Z-scores are an essential tool for comparing different distributions because they standardize data, making it possible to compare values from different distributions, even if those distributions have different scales, means, and standard deviations. This helps us assess the relative position of data points or sample means within different distributions.

21) What are the assumptions for applying the Central Limit Theorem ?
>   The Central Limit Theorem (CLT) is a fundamental principle in statistics, but like many powerful tools, it comes with certain assumptions. Here are the key assumptions you need to consider when applying the CLT.
Assumptions: a) Independence: The individual observations must be independent of each other.
b) Identically Distributed: The observations should be identically distributed, meaning they are all drawn from the same probability distribution with a common mean and variance.
c) Sample Size: The sample size should be sufficiently large.A common rule of thumb is that a sample size of 30 or more
d) Finite Variance: The population from which the samples are drawn should have a finite variance. If the population variance is infinite, the CLT does not apply.

22) What is the concept of expected value in a probability distribution ?
>   The expected value (EV), also known as the mean or mathematical expectation, is a fundamental concept in probability and statistics. It represents the average or central tendency of a random variable, providing a long-term prediction of its value. Essentially, it tells you what you would expect as an "average" outcome if an experiment were repeated many times.

23) How does a probability distribution relate to the expected outcome of a random variable ?
>   A probability distribution describes the likelihood of all possible outcomes of a random variable. It gives us a complete picture of how the values of a random variable are distributed, showing the probabilities associated with each possible outcome. The expected outcome (or expected value) of a random variable is derived from its probability distribution and represents the "average" or central tendency of that distribution.

```python
#3) Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculate the Probability Distribution Function (PDF) of a Bernoulli distribution.
    
    Args:
    x (int): The value of the random variable (0 or 1).
    p (float): The probability of success (0 <= p <= 1).
    
    Returns:
    float: The probability of observing x for the Bernoulli distribution.
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1 for Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1 inclusive.")
    
    # Bernoulli PDF calculation
    if x == 1:
        return p
    else:
        return 1 - p

p = 0.7  # Probability of success
x_1 = 1  # Outcome for success
x_0 = 0  # Outcome for failure

print(f"P(X = {x_1}) = {bernoulli_pdf(x_1, p)}")
print(f"P(X = {x_0}) = {bernoulli_pdf(x_0, p)}")
```

```python
#9) Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    
    mean = np.mean(data)
    std_dev = np.std(data)
    
    z_scores = (data - mean) / std_dev
    
    
    plt.figure(figsize=(8, 6))
    plt.plot(z_scores, marker='o', linestyle='-', color='skyblue', markersize=5, label='Z-scores')
    plt.title('Z-Scores of the Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.axhline(0, color='black', linewidth=1, linestyle='--')  # Add a line at Z=0 for reference
    plt.grid(True)
    plt.legend()
    plt.show()
    
    return z_scores

data = np.random.normal(0, 1, 1000)  # Generate a normal distribution with mean=0, std=1, and 1000 samples
z_scores = calculate_and_plot_z_scores(data)
```

```python
#10) Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem(num_samples, sample_size, low, high):
    sample_means = []
    
    for _ in range(num_samples):
        sample = np.random.uniform(low, high, sample_size)
        sample_means.append(np.mean(sample))
    
    sample_means = np.array(sample_means)
    
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(f'Central Limit Theorem: Sample Means of Non-Normal Distribution')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()
    
    return sample_means

num_samples = 1000 
sample_size = 30   
low = 0             
high = 10            

sample_means = central_limit_theorem(num_samples, sample_size, low, high)
```

```python
#13)  Generate random variables and calculate their corresponding probabilities using the binomial distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def binomial_distribution(n, p, num_trials):
    
    random_vars = np.random.binomial(n, p, num_trials)
    probabilities = binom.pmf(random_vars, n, p)
    
    plt.hist(random_vars, bins=range(n+2), density=True, alpha=0.7, color='skyblue', edgecolor='black', align='left')
    
    x = np.arange(0, n+1)
    pmf_values = binom.pmf(x, n, p)
    plt.plot(x, pmf_values, 'ro', label='PMF', color='red')
    
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return random_vars, probabilities

n = 10             # Number of trials
p = 0.5            # Probability of success on each trial
num_trials = 1000  # Number of random variables to generate

random_vars, probabilities = binomial_distribution(n, p, num_trials)

print("Random Variables and Corresponding Probabilities:")
for rv, prob in zip(random_vars[:10], probabilities[:10]):
    print(f"Random Variable: {rv}, Probability: {prob:.4f}")
```

```python
#14)  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
#distribution.

import scipy.stats as stats

def calculate_z_score(data_point, mean, std_dev):
    # Formula for Z-score
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_normal_distribution(z_score):
    # Compare the Z-score with the standard normal distribution
    # P-value from the Z-score using the CDF (Cumulative Distribution Function)
    p_value = stats.norm.cdf(z_score)
    
    if p_value > 0.95:
        return f"The data point is in the upper 5% of the distribution (p-value: {p_value:.4f})"
    elif p_value < 0.05:
        return f"The data point is in the lower 5% of the distribution (p-value: {p_value:.4f})"
    else:
        return f"The data point is within the middle 90% of the distribution (p-value: {p_value:.4f})"

data_point = float(input("Enter the data point: "))
mean = float(input("Enter the mean of the distribution: "))
std_dev = float(input("Enter the standard deviation of the distribution: "))

# Calculate Z-score
z_score = calculate_z_score(data_point, mean, std_dev)
print(f"The Z-score is: {z_score:.2f}")

# Compare to normal distribution
result = compare_to_normal_distribution(z_score)
print(result)
```

```python
#15) Implement hypothesis testing using Z-statistics for a sample dataset.

import scipy.stats as stats
import math

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    # Step 1: Calculate sample mean and sample size
    sample_mean = sum(sample_data) / len(sample_data)
    n = len(sample_data)

    # Step 2: Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / math.sqrt(n))

    # Step 3: Find the critical Z-value for the two-tailed test
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Step 4: Compare the calculated Z-statistic with critical value
    if abs(z_statistic) > z_critical:
        result = "Reject the null hypothesis: The sample mean is significantly different from the population mean."
    else:
        result = "Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean."

    return z_statistic, result

sample_data = [150, 160, 145, 155, 162, 148, 155, 160, 159, 151]
population_mean = 150
population_std = 8

z_statistic, result = z_test(sample_data, population_mean, population_std)
print(f"Calculated Z-statistic: {z_statistic:.2f}")
print(result)
```

```python
#23 Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate dataset
data = np.random.normal(10, 2, 1000)  # mean of 10, std dev of 2

# Implement Z-score Normalization
mean = np.mean(data)
std_dev = np.std(data)
normalized_data = (data - mean) / std_dev

# Visualize the Transformation
sns.set(style="whitegrid")

# Plot original and normalized data
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(data, kde=True, color="blue")
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
sns.histplot(normalized_data, kde=True, color="green")
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#24 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#distribution.

import numpy as np
from scipy.stats import skew, kurtosis

mean = 0
std_dev = 1
sample_size = 1000

data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

# Define functions to calculate skewness and kurtosis
def calculate_skewness(data):
    return skew(data)

def calculate_kurtosis(data):
    return kurtosis(data)

# Execute the functions
data_skewness = calculate_skewness(data)
data_kurtosis = calculate_kurtosis(data)

print(f"Skewness: {data_skewness}")
print(f"Kurtosis: {data_kurtosis}")
```
